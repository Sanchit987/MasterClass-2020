```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
import networkx as nx
```

```python
G_karate = nx.karate_club_graph()
```

# Find number of nodes and edges in the graph

```python
print(nx.info(G_karate))
```

# find betweenness centrality of all nodes and visualise top 5 nodes with highest betweenness centrality

```python
bc = nx.betweenness_centrality(G_karate)
top = 5

max_nodes =  sorted(bc.items(), key = lambda v: -v[1])[:top]
bt_values = [50]*len(G_karate.nodes())
bt_colors = [0]*len(G_karate.nodes())
for node_deg in max_nodes:
    max_key=list(G_karate.nodes()).index(node_deg[0])
    bt_values[max_key] = 150
    bt_colors[max_key] = 2
```

```python
plt.figure(figsize=(15,10))
nx.draw_networkx(G_karate, cmap = plt.get_cmap("rainbow"), node_color = bt_colors, node_size = bt_values, with_labels = True)
```

# # find degree centrality of all nodes and visualise top 5 nodes with highest degree centrality

```python
dc = nx.degree_centrality(G_karate)
top = 10

max_nodes =  sorted(dc.items(), key = lambda v: -v[1])[:top]
dc_values = [5]*len(G_karate.nodes())
dc_colors = [0]*len(G_karate.nodes())
for node_deg in max_nodes:
    max_key=list(G_karate.nodes()).index(node_deg[0])
    dc_values[max_key] = 150
    dc_colors[max_key] = 2
    
```

```python
plt.figure(figsize=(15,10))
pos = nx.spring_layout(G_karate)
nx.draw_networkx(G_karate, pos = pos, cmap = plt.get_cmap("rainbow"), node_color = dc_colors,
                 node_size = dc_values, with_labels = True)
```
