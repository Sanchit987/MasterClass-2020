```
import tensorflow.compat.v1 as tf
```

Creating a **Static** Graph

A limitation of TensorFlow 0.x

```
mygraph = tf.Graph()
print('AN empty graph is created')
```

```
with mygraph.as_default():
  x = tf.placeholder(tf.float32,shape=[None])
  y = tf.placeholder(tf.float32,shape=[None])
  print('Placeholders Created')
```

Evaluate Z = ( X + Y ) - ( X * Y )

***Data = Nodes***

***Operations = Edges***

Note which of these operations can be done in parallel.

Operators are overloaded

```
with mygraph.as_default():
  node_plus = x + y
  node_mult = x * y
  node_mins = node_mult - node_plus
  print('Operations defined')
```

Create / Load data (Numpy array) to feed the Graph

```
import numpy as np

x_data = np.random.randint(0,50,[5])
y_data = np.random.randint(0,30,[5])
print(x_data)
print(y_data)
```

```
with tf.Session(graph=mygraph) as sess:
  feed_data = {x:x_data,y:y_data}
  result = sess.run([node_mins],feed_dict=feed_data)
  print(result)
```
